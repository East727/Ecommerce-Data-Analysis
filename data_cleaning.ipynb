```python
import numpy as np
import pandas as pd

# 1.读取数据
df = pd.read_csv('../data/OnlineRetail.csv',encoding='latin-1')

# 2.查看基本信息
# print(df.info())
# print('数据形状：',df.shape)
print(df.head())
```

```python
# 3.数据清洗
# 查看缺失值
# print(df.isnull().sum())
# print(df.isna().sum())

# 查看重复值
# print(df.duplicated())
# print(df.duplicated().sum())
# print(df.count())

# 去重
# df.drop_duplicates()
# print(df.duplicated().sum())
print(df.shape[0])
```

```python
# 4.数据处理, 删除多余行
# 删除 CustomerID 为空的记录
df = df.dropna(subset=['CustomerID'])

# 删除数量 <= 0 或单价 <= 0 的记录（退货或无效数据）
df = df[(df['Quantity']>0) & (df['UnitPrice']>0)]
```

```python
# print(df.shape)
# 5. 新增列： 销售额 = 数量 * 单价
df['TotalAmount'] = df['Quantity'] * df['UnitPrice']
# print(df.head(2))
```

```python
# 6.提取日期列（年、月、日）
df['InvoiceDate'] = pd.to_datetime(df['InvoiceDate'])
df['Year'] = df['InvoiceDate'].dt.year
df['Month'] = df['InvoiceDate'].dt.month
df['Day'] = df['InvoiceDate'].dt.day
# print(df.head(2))
print(df.info())
```

```python
# 7.保存数据
df.to_csv('../data/OnlineRetail_clean.csv',index=False)
print('数据保存成功！')
```
